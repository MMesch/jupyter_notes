```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from scipy.stats import laplace

npoints, nfeatures = 500, 200

freq1 = 10.
freq2 = 20
signal = np.random.normal(scale=0.5, loc=0, size=(npoints * nfeatures))
signal = signal.reshape(npoints, nfeatures)
signal += np.sin(2 * np.pi * freq1 * np.linspace(0., 1, nfeatures))[None, :] * laplace.rvs(size=npoints)[:, None]
signal += np.sin(2 * np.pi * freq2 * np.linspace(0., 1, nfeatures))[None, :] * laplace.rvs(size=npoints)[:, None]

fig, ax = plt.subplots(figsize=(10, 30))
ax.imshow(signal, aspect='auto', cmap='RdBu');
ax.set(xlabel='features', ylabel='samples');
```

```python
sns.clustermap(signal, metric='correlation', col_cluster=False, method='complete');
```
